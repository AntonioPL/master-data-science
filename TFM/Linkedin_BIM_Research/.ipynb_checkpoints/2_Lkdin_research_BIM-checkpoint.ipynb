# [Importing and loading files]

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as axes
import string
from collections import deque
%matplotlib inline
plt.style.use('ggplot')
```

```python
file='./DBs/Linkedin_searches.csv'
file_updated='./DBs/Li_research_updated_1.csv'
rcsv = pd.read_csv(file)
rcsvu= pd.read_csv(file_updated,encoding = "ISO-8859-1")
rcsvu
```

# [DFs declaration]

```python
#Let's declare working DFs and fill new aggregated entries about the total amoun of BIM profiles 
#at years 2015 and 2016 according to consultants information: 
bim_stats= rcsvu.loc[[21],['BIM_Role','Unprecise_role 2017','Unprecise_role 2018']]
bim_stats=bim_stats.transpose()
bim_stats=bim_stats.drop(['BIM_Role'])
bim2015= pd.DataFrame({'year': 'Unprecise_role 2015', 'hits': [4658]})
bim2016= pd.DataFrame({'year': 'Unprecise_role 2016', 'hits': [5892]})

roles=rcsvu.loc[:,['BIM_Role','Exact_role 2017','Unprecise_role 2017','Exact_role 2018','Unprecise_role 2018','Perc']].drop([21])
bim_stats.columns = ['hits']
bim_stats['year'] = bim_stats.index

bim_stats=bim2015.append([bim2016,bim_stats]).reset_index()
del bim_stats['index']

bim_stats=bim_stats.sort_values('hits',ascending=False)
roles=roles.sort_values('Exact_role 2018',ascending=True)

#Creamos Columna con roles y su número de orden.

def creaordercol(df):
    i=0
    newserie = pd.Series()
    alph = list(string.ascii_uppercase)
    for row in df.itertuples():      
        newserie=newserie.append(pd.Series([alph[i]+'.'+ str(row.BIM_Role)]))
        i+=1
    return newserie

BIM_rol_ord = creaordercol(roles)


roles_ord =roles.assign(BIM_rol_ord=BIM_rol_ord.values)
roles_ord


```

La línea de mínimos ploteada expresa la línea base desde la cual el sector de la construcción español comenzaría a trabajar en un ecosistema BIM sano y fluído. Se requieren muchas sinergias para consolidar buenas rutinas entre profesionales, empresas y clientes con estas nuevas metodologías; y el mercado laboral (en cuanto a cantidad y calidad de profesionales) puede ser la llave o el cuello de botella si no se le da suficiente consideración.

# Categorization of the aggregated BIM roles and the evolution of the nomenclature.

Conclusions: There are few BIM roles consolidated for professionals and production teams. As we can see, there are many declared BIM Managers, but few BIM Modelers, developers and engineering specialists. The first conclussion is that many professionals apply directly to BIM Management roles without having enough previous production experience.

Another topic to take into account is the diversification or roles (BIM consultant,BIM drawing specialist,...). It is partially related with the previous observation about BIM Management role. As we can see, between 2017 and 2018 people labeled as "BIM manager" changed to another role. The most probable explanation is that most of BIM professionals start studying a Master or come from a construction management role. In their situation, as novices in this new field, they would only know about "BIM Manager" role. This, plus the aparent convenience of beginning as a "manager" in this field, push these professionals to label themselves as "BIM Managers" in social networks until other more feasible roles are applied to them form real market.  

Nevertheless, the visible changes in the quantities and the new roles are remarkable and are, somehow, natural in a field that is inmerse in hard trial and error routines.

One last observation is about those who have formation or real experience in BIM managment. Due to the complete lack of professionals in every role but management related, it is possible that there are many underrated professionals developing too-easy jobs for them. It is very probable that they use their know-how in their everyday job, even if it is not explicitly required, as the current BIM enviroment has a chronic lack of BIM protocols and standards. This is the optimal professional for BIM Management role, and our job is find those "hidden" underrated professionals capable of creating new production structures.

# Categorización de la masa de roles BIM y la normalización de su nomenclatura.

Conclusiones: Hay pocos roles BIM consolidados entre profesionales y equipos de producción. Como podemos observar, hay muchos BIM Managers declarados y muy pocos modeladores BIM, desarrolladores de aplicaciones e ingenieros especialistas. La primera conclusión que sacamos de este fenómeno es que muchos professionales optan a roles de Gestión BIM sin tener suficiente experiencia previa.

Otro tema a tener en cuenta es la diversificación de roles (consultor BIM, delineante BIM,...). Este tema está parcialmente relacionado con la anterior observación sobre el rol de Gestor BIM. Como podemos ver, entre 2017 y 2018, se redujo la cantidad de BIM Managers, los cuales se cambiarían a otro rol de empresa. La explicación más probable es que la mayoría de professionales BIM comienzan estudiando un Master y/o provienen de un perfil de gestión de la construcción. En su situación, como recién llegados a este nuevo campo, solo conocerían el perfil de "gestor BIM". Esto, sumado a la conveniencia aparente de comenzar commo "gestor" en este campo, empuja a los profesionales a etiquetarse como "Gestores BIM" o "BIM Manager" en las redes sociales hasta que otro rol más apropiado les sea otorgado por el mercado laboral.

En cualquie caso, los cambios en las cantidades y los nuevos roles son destacables y, por otro lado, naturales en un campo que está inmerso en complicadas tareas de prueba y error.

Una última observación relativa a los gestores BIM que tienen formación o incluso experiencia real. Dada la alta necesidad del mercado en otros perfiles que no son de gestión, hay una alta probabilidad de que haya una notable cantidad de profasionales infra Por lo que cuando se emplean dejan de ejercer el rol de BIM Manager en sentido estricto, aunque no dejen de usar sus aptitudes en el campo

```python
# Barplotting of evolution of BIM roles:

roles_graph = roles_ord.set_index('BIM_rol_ord')
fig = plt.figure(figsize=(17,8))
plt.xlabel("BIM Roles according to online professional profiles")
plt.ylabel("Number of exact coincidences")
plt.title("BIM ROLES DISTRIBUTION AT 2017 & 2018 \n")
plt.xticks(rotation=70)

x=roles_graph.index.values
y1=roles_ord['Exact_role 2017']
y2=roles_ord['Exact_role 2018']

pl1= plt.bar(x,y1, width=-0.30, align='edge',color='#BAC6FF')
pl2=plt.bar(x,y2, width=0.3, align='edge',color='#5F7AFF')

plt.legend([pl1, pl2], ["Profile search in 2017", "Profile search in 2018"],loc=0,fontsize=12,
          bbox_to_anchor=(0., 1.02, 1., .102),
           ncol=1, borderaxespad=0.)

plt.show()
```

```python
roles_perc = roles_ord['Perc']
roles_demand  = pd.concat([roles_ord['BIM_Role'],roles_ord['Exact_role 2018'], roles_perc],axis=1)

#Ordenamos de nuevo las categorías según los roles requeridos 'Perc':
alph = list(string.ascii_uppercase)

bim_stats_demand = roles_demand.sort_values('Perc',ascending=False)

alph = deque(list(string.ascii_uppercase))

def alphab(stri):
    return alph.popleft()+'.'+stri
        
        
bim_stats_demand['BIM_Role']=bim_stats_demand['BIM_Role'].apply(lambda x: alphab(x))
bim_stats_demand
```

# [ESP] Estudio comparativo de balance de oferta y demanda de perfiles profesionales en el mercado.



Pese al evidente aumento de perfiles BIM observado en el apartado de "progresión del concepto BIM en el mercado laboral", destaca la escasez de perfiles de producción y coordinación de modelos BIM así como de técnicos más orientados al desarrollo de aplicaciones y análisis de la información de los modelos. En concreto escasea personal con experiencia en modelado 3d complejo (en su vertiente geométrica y de datos bien estructurados) y en programación de aplicaciones de automatización de comandos.

```python
# Preparación del DF con nueva columna de perfiles demandados:

roles_demand  = pd.concat([roles_ord['BIM_Role'],roles_ord['Exact_role 2018'], roles_ord['Perc']],axis=1)

#Ordenamos de nuevo las categorías según los roles requeridos 'Perc':
bim_stats_demand = roles_demand.sort_values('Perc',ascending=True)

#Función asignar letra para orden de categoría en el eje.
alph = deque(list(string.ascii_uppercase))
def alphab(stri):
    return alph.popleft()+'.'+stri        
        
bim_stats_demand['BIM_Role']=bim_stats_demand['BIM_Role'].apply(lambda x: alphab(x))
bim_stats_demand

#Barplotting BIM role grading in relation to the demand of the labour market.

fig = plt.figure(figsize=(17,8))
plt.xlabel("Roles BIM según perfiles profesionales Online")
plt.ylabel("Puntuación oferta / demanda")
plt.title("BALANCE OFERTA / DEMANDA \nDE PERFILES PROFESIONALES en 2018 \n")
plt.xticks(rotation=70)

x=bim_stats_demand['BIM_Role']
y1=bim_stats_demand['Exact_role 2018']*0.016
y2=bim_stats_demand['Perc']*0.016

pl2=plt.bar(x,y2, width=0.25, align='edge',color='#ff6666')
pl1= plt.bar(x,y1, width=-0.25, align='edge',color='#3498db')


plt.legend([pl1, pl2], ["Perfiles existentes año 2018", "Demanda de perfiles según medianas \nen equipos de producción"],loc=0,fontsize=12,
          bbox_to_anchor=(0., 1.02, 1., .102),
           ncol=1, borderaxespad=0.)

plt.show()
```
